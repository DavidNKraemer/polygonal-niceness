```python
chord_area_score_data = np.loadtxt('../data/regular_ngon_chord_area_score.txt')

ngons = np.arange(3,3+chord_area_score_data.size,1)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,6))
ax.plot(ngons,chord_area_score_data,'k--o')
ax.set_title('Chord-area score for unit-area regular $n$-gon')
ax.set_xlabel('$n$')
ax.set_ylabel('Chord-area score');
```
